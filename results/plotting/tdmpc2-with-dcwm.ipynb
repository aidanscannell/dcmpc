# TD-MPC2 with DCWM (Figure 6)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import utils
from rliable import library as rly
from rliable import metrics, plot_utils
```

```python
plt.rcParams["font.size"] = 18
plt.rcParams["legend.fontsize"] = 18
```

```python
COLORS = {
    "TD-MPC2 w/ DCWM": "#e41a1c",
    "TD-MPC2": "#377eb8",
    "DC-MPC (ours)": "#984ea3",
}
AGENTS = {
    "TD-MPC2 w/ FSQ": "TD-MPC2 w/ DCWM",
    "TD-MPC2": "TD-MPC2",
    "DCWM": "DC-MPC (ours)",
}

DMC_ENVS = [
    "acrobot-swingup",
    "dog-run",
    "dog-stand",
    "dog-trot",
    "dog-walk",
    "humanoid-run",
    "humanoid-stand",
    "humanoid-walk",
    "reacher-hard",
    "walker-walk",
]
MW_ENVS = [
    "mw-button-press",
    "mw-door-open",
    "mw-drawer-close",
    "mw-drawer-open",
    "mw-peg-insert-side",
    "mw-pick-place",
    "mw-push",
    "mw-reach",
    "mw-window-open",
    "mw-window-close",
]
```

```python
df = pd.read_csv("../data/ablations/tdmpc2-with-dcwm.csv")
```

```python
def plot_interval_estimates(df, verbose: bool = True):
    def check_envs():
        """Checks what seeds each agent has in each env"""
        for agent in df.agent.unique():
            for env in df.env.unique():
                agent_envs = df[df["agent"] == agent].env.unique()
                if env not in agent_envs:
                    print(f"{env} is NOT in {agent}")
                num_seeds = df[
                    (df["agent"] == agent) & (df["env"] == env)
                ].seed.unique()
                env_steps = df[
                    (df["agent"] == agent) & (df["env"] == env)
                ].env_step.unique()
                print(
                    f"seeds={np.sort(num_seeds)} for {agent} in {env}, env_step={env_steps}"
                )

    if verbose:
        check_envs()

    # Initialize an empty dictionary to store arrays for each agent
    agent_arrays = {}

    # Loop over each unique agent to create a separate array for each
    for agent in df["agent"].unique():
        # Filter the DataFrame for the current agent
        agent_df = df[df["agent"] == agent]

        # Pivot the filtered DataFrame to create a multi-indexed table with "seed" as rows, "env" and "env_step" as columns
        pivoted_df = agent_df.pivot_table(
            values="episode_reward", index="seed", columns=["env", "env_step"]
        )

        # Convert the pivoted DataFrame to a numpy array
        array = pivoted_df.to_numpy(copy=True)

        # # Normalize the scores to be in 0-1
        # array = array / max_score

        # Reshape the array based on the unique counts of "seed", "env", and "env_step"
        array = array.reshape(
            len(agent_df["seed"].unique()),
            len(agent_df["env"].unique()),
            len(agent_df["env_step"].unique()),
        )

        # Add the array to the dictionary with the agent as the key
        agent_arrays[agent] = array

    scores_dict = agent_arrays

    # Load ALE scores as a dictionary mapping algorithms to their human normalized
    # score matrices, each of which is of size `(num_runs x num_games)`.
    # atari_200m_normalized_score_dict = ...
    aggregate_func = lambda x: np.array(
        [
            metrics.aggregate_iqm(x),
            metrics.aggregate_optimality_gap(x),
        ]
    )
    aggregate_scores, aggregate_score_cis = rly.get_interval_estimates(
        scores_dict, aggregate_func, reps=50000
    )

    fig, axes = plot_utils.plot_interval_estimates(
        aggregate_scores,
        aggregate_score_cis,
        metric_names=[
            "IQM",
            "Optimality Gap",
        ],
        xlabel="Normalized Score",
        colors=COLORS,
        xlabel_y_coordinate=-0.3,
        row_height=0.7,
    )
    plt.subplots_adjust(wspace=0.2)
    return axes


def plot_tdmpc2_with_dcwm_interval_estimates(df, verbose: bool = True):
    env_step = 1000000

    # Only keep desired agents
    df = df[df["agent"].isin(AGENTS.keys())]
    df["agent"] = df["agent"].replace(AGENTS)

    # Make sure we have seed=1,2,3 for each agent/env
    SEEDS = [1, 2, 3]
    df = df[df["seed"].isin(SEEDS)]

    # Only keep desired environments
    df_dmc = df[df["env"].isin(DMC_ENVS)]
    df_dmc = df_dmc[df_dmc["env_step"] == env_step]
    df_dmc["episode_reward"] = df_dmc.apply(
        lambda row: row["episode_reward"] / 1000, axis=1
    )

    # Prepare Meta-World data frame
    df_mw = df[df["env"].isin(MW_ENVS)]
    df_mw = df_mw[df_mw["env_step"] == env_step]
    df_mw["episode_reward"] = df_mw["episode_success"]
    df_normalized = pd.concat([df_dmc, df_mw])

    axs = plot_interval_estimates(df_normalized, verbose=verbose)
    # axs[0].set_xticks((0.8, 0.85, 0.9, 0.95))
    axs[0].set_xticks((0.8, 0.9, 1.0))
    axs[1].set_xticks((0.15, 0.2, 0.25))

    # plt.tight_layout()
    plt.savefig(f"./figs/tdmpc2-with-dcwm-iqm-optimality-gap.pdf", bbox_inches="tight")

plot_tdmpc2_with_dcwm_interval_estimates(df, verbose=False)
```

```python
def plot_iqm(
    df, env_steps, env_name, max_score=None, legend=False, key="episode_reward", ax=None, verbose=True
):
    if max_score is None:
        max_score = df.episode_reward.max()

    def check_envs():
        """Checks what seeds each agent has in each env"""
        for agent in df.agent.unique():
            for env in df.env.unique():
                agent_envs = df[df["agent"] == agent].env.unique()
                if env not in agent_envs:
                    print(f"{env} is NOT in {agent}")
                num_seeds = df[
                    (df["agent"] == agent) & (df["env"] == env)
                ].seed.unique()
                env_steps = df[
                    (df["agent"] == agent) & (df["env"] == env)
                ].env_step.unique()
                print(
                    f"seeds={np.sort(num_seeds)} for {agent} in {env}, env_step={env_steps} len(env_step)={len(env_steps)}"
                )

    if verbose:
        check_envs()

    # Initialize an empty dictionary to store arrays for each agent
    agent_arrays = {}

    # Loop over each unique agent to create a separate array for each
    for agent in df["agent"].unique():
        # Filter the DataFrame for the current agent
        agent_df = df[df["agent"] == agent]

        # Pivot the filtered DataFrame to create a multi-indexed table with "seed" as rows, "env" and "env_step" as columns
        pivoted_df = agent_df.pivot_table(
            values=key, index="seed", columns=["env", "env_step"]
        )

        # Convert the pivoted DataFrame to a numpy array
        array = pivoted_df.to_numpy(copy=True)

        # Normalize the scores to be in 0-1
        array = array / max_score

        # Reshape the array based on the unique counts of "seed", "env", and "env_step"
        array = array.reshape(
            len(agent_df["seed"].unique()),
            len(agent_df["env"].unique()),
            len(agent_df["env_step"].unique()),
        )

        # Add the array to the dictionary with the agent as the key
        agent_arrays[agent] = array

    scores_dict = agent_arrays

    iqm = lambda scores: np.array(
        [metrics.aggregate_iqm(scores[..., frame]) for frame in range(scores.shape[-1])]
    )
    iqm_scores, iqm_cis = rly.get_interval_estimates(
        scores_dict,
        iqm,
        reps=50000,
        task_bootstrap=False,
    )
    ax = plot_utils.plot_sample_efficiency_curve(
        env_steps + 1,
        iqm_scores,
        iqm_cis,
        colors=COLORS,
        xlabel=r"Env. Steps (1e3)",
        ylabel="",
        legend=legend,
        ax=ax,
    )

    ax.set_xticks((0, 0.5e6, 1e6))
    ax.set_xticklabels(("0", "500", "1000"))

    num_tasks = len(df.env.unique())
    ax.set_title(f"{env_name} {num_tasks} tasks",fontsize=30)
    return ax


def plot_tdmpc2_with_dcwm_sample_eff(df, verbose: bool = True):
    # only keep desired agents
    df = df[df["agent"].isin(AGENTS.keys())]
    df["agent"] = df["agent"].replace(AGENTS)

    ncol = 2
    nrow = 1
    fig, axs = plt.subplots(
        nrow,
        ncol,
        figsize=(5 * ncol, 3.8 * nrow),
        sharey="row",
    )

    # scale episode by 1000
    df["episode"] = df.apply(lambda row: int(row["env_step"] / 1000), axis=1)

    # make sure we have seed=1,2,3 for each agent/env
    seeds = [1, 2, 3]
    df = df[df["seed"].isin(seeds)]

    env_steps = np.arange(0, 11) * 100000

    # only keep desired environments
    df_dmc = df[df["env"].isin(DMC_ENVS)]
    df_dmc = df_dmc[df_dmc["env_step"].isin(env_steps)]

    # prepare meta-world data frame
    df_mw = df[df["env"].isin(MW_ENVS)]
    df_mw = df_mw[df_mw["env_step"].isin(env_steps)]

    env_names = ["DMControl", "Meta-World"]
    keys = ["episode_reward", "episode_success"]
    max_scores = [1000, 1]
    for i, df_env in enumerate([df_dmc, df_mw]):
        ax = plot_iqm(
            df_env,
            env_steps,
            env_name=env_names[i],
            max_score=max_scores[i],
            key=keys[i],
            ax=axs[i],
            verbose=verbose,
        )

    axs[0].set_ylabel("IQM Normalized Score")

    plt.savefig(f"./figs/tdmpc2-with-dcwm-sample-eff.pdf", bbox_inches="tight")

plot_tdmpc2_with_dcwm_sample_eff(df, verbose=False)
```
