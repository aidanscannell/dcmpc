# Ablation of FSQ vs VQ (Figure 10)

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
import seaborn as sns
import utils
```

```python
plt.rcParams["font.size"] = 19
plt.rcParams["legend.fontsize"] = 19
```

```python
ENCODINGS = {
    "codes": "FSQ",
    "vq-weighted-avg": "VQ (Weighted Avg)",
    "vq-sample": "VQ (Sample)",
}
PALETTE = {
    "FSQ": "#984ea3",
    "VQ (Weighted Avg)": "#377eb8",
    "VQ (Sample)": "#e41a1c",
}
```

```python
df = pd.read_csv("../data/ablations/fsq-vq.csv")
```

```python
# Convert env step to (1e3)
df.loc[:, "env_step"] = df["env_step"] // 1000
```

```python
def plot_fsq_vq_ablation(df, key="episode_reward"):
    ncol = 4
    nrow = 1
    fig, axs = plt.subplots(
        nrow,
        ncol,
        figsize=(3 * ncol, 3.5 * nrow),
        sharex="col",
        sharey="row",
    )

    df["discrete_encoding_type"] = df["discrete_encoding_type"].replace(ENCODINGS)

    for idx, env in enumerate(np.sort(df.env.unique())):
        data = df[df["env"] == env]
        col = idx % ncol
        ax = axs[col]
        hue_order = np.sort(data.discrete_encoding_type.unique())
        legend = idx == len(df.env.unique()) - 1

        g = sns.lineplot(
            x="env_step",
            y=key,
            data=data,
            errorbar=("ci", 95),
            hue="discrete_encoding_type",
            hue_order=hue_order,
            palette=PALETTE,
            legend="auto" if legend else False,
            ax=ax,
        )

        ax.set_title(" ".join([ele.capitalize() for ele in env.split("-")]))
        ax.set_xlabel("Env. Steps (1e3)")
        ax.set_ylabel("")
        if env in ["walker-walk"]:
            g.set(xlim=(0, 250))
        else:
            g.set(xlim=(0, 1500))
    axs[0].set_ylabel(utils.YLABELS[key])

    l = axs[-1].legend(
        loc="upper center",
        bbox_to_anchor=(-1.5, -0.35),
        fancybox=False,
        shadow=False,
        ncol=len(df.discrete_encoding_type.unique()),
    )
    for line in l.get_lines():
        line.set_linewidth(5.0)
    # plt.tight_layout()
    fig.subplots_adjust(
        bottom=0.35,
        hspace=0.3,
        wspace=0.3,
        left=0.08,
        right=0.97,
        top=0.88,
    )
    plt.savefig(f"./figs/fsq-vq-ablation.pdf")

plot_fsq_vq_ablation(df)
```
